# Functions

A **function** is a fundamental concept in mathematics that describes a relationship between two sets, typically called the *domain* and *codomain*. In simple terms, a function assigns each element from the domain to exactly one element in the codomain. The key idea is that every input has a unique output.

Mathematically, a function $f$ is often denoted as follows:

$$\large f: A \rightarrow B $$

Here:
- $ f $ represents the function.
- $ A $ is the *domain*, which is the set of all possible input values.
- $ B $ is the *codomain*, which is the set of all possible output values.

For each $ x $ in $ A $, there exists a unique $ y $ in $ B $ such that $ f(x) = y $. The set of all possible $ y $ values that the function can produce is called the *range*.

## Domain and Codomain:

- **Domain:** This is the set of all possible input values for a function. It's the "input space" for the function. In the notation $ f: A \rightarrow B $, $ A $ is the *domain*.

- **Codomain:** This is the set of all possible output values. It represents the "output space" for the function. In the notation $ f: A \rightarrow B $, $ B $ is the *codomain*.

It's important to note that not every element in the codomain needs to have a corresponding element in the domain. The elements of the codomain are simply the possible values that the function could output.

## Explicit and Implicit Forms:

- **Explicit Form:** In an explicit form, a function is defined by an algebraic expression that directly relates the input to the output. For example, the function $ f(x) = 2x + 3 $ is in explicit form. You can easily see how the input ($ x $) is transformed into the output.

- **Implicit Form:** In an implicit form, the relationship between the input and output may not be immediately apparent. Equations involving both the input and output variables can define functions implicitly. For example, the equation $ x^2 + y^2 = 1 $ implicitly defines a function because for each $ x $, there are two possible $ y $ values (considering both positive and negative square roots). Implicitly defined functions are often represented by equations involving both variables.

Understanding these concepts is crucial in various mathematical applications, including calculus, linear algebra, and real-world problem-solving. Functions provide a way to model and understand relationships between quantities and are fundamental in many areas of mathematics and science.


## Studying the Sign of a Function:

When studying the sign of a function $f(x)$, we are interested in determining where the function is positive, negative, or zero. This information is crucial for understanding the behavior of the function.

### Steps:

1. **Find Critical Points:**
   - Critical points occur where the function is equal to zero or undefined. Set $f(x) = 0$ and find any values of $x$ that make the function undefined.

2. **Test Intervals:**
   - Use test points from each interval determined by the critical points to evaluate the sign of the function. Choose a point in each interval and substitute it into the function. The sign of the result indicates the sign of the function in that interval.

3. **Determine Sign:**
   - Based on the evaluations, determine where the function is positive, negative, or zero.

## Calculating the Domain of a Function:

The domain of a function $f(x)$ is the set of all possible input values (values of $x$) for which the function is defined.

### Steps:

1. **Identify Restrictions:**
   - Look for any values of $x$ that make the function undefined. Common restrictions include division by zero and square roots of negative numbers.

2. **Express the Domain:**
   - Express the domain using mathematical notation. For example, if the function is defined for all real numbers, the domain is $\mathbb{R}$. If there are restrictions, express them explicitly.

## Example (in code):

Let's consider the function $f(x) = \frac{1}{x^2 - 4}$:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return 1 / (x**2 - 4)

# Generate x values
x_values = np.linspace(-5, 5, 400)
# Avoid division by zero by excluding -2 and 2 from the domain
x_values = x_values[(x_values < -2) | (x_values > 2)]

# Generate y values
y_values = f(x_values)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$f(x) = \frac{1}{x^2 - 4}$', color='blue')

# Set the x and y limits
plt.xlim(-5, 5)
plt.ylim(-5, 10)

# Highlight critical points
plt.scatter([-2, 2], [0, 0], color='red', marker='o', label='Critical Points')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of the Function')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Show the plot
plt.show()
```
