# Functions

A **function** is a fundamental concept in mathematics that describes a relationship between two sets, typically called the *domain* and *codomain*. In simple terms, a function assigns each element from the domain to exactly one element in the codomain. The key idea is that every input has a unique output.

Mathematically, a function $f$ is often denoted as follows:

$$\large f: X \rightarrow Y $$

Here:
- $ f $ represents the function.
- $ X $ is the *domain*, which is the set of all possible input values.
- $ Y $ is the *codomain*, which is the set of all possible output values.

For each $ x $ in $ X $, there exists a unique $ y $ in $ Y $ such that $ f(x) = y $. The set of all possible $ y $ values that the function can produce is called the *range*.

It can happen that the same y is the image of different points in $X$, that is, we have $y = f(x) = f(x₀)$ for $x, x₀ ∈ X$, and $x ≠ x₀$; what cannot happen is that for an $x ∈ X$, two distinct elements of $Y$ are associated, that is, resulting in $f(x) = y$ and $f(x) = y₀$ with $y ≠ y_0$.

The input to the function is called the independent variable, and is also called the argument of the function. The output of the function is called the dependent variable.

## Domain and Codomain

- **Domain:** This is the set of all possible input values for a function. It's the "input space" for the function. In the notation $ f: X \rightarrow Y $, $ X $ is the *domain*.

- **Codomain:** This is the set of all possible output values. It represents the "output space" for the function. In the notation $ f: X \rightarrow Y $, $ Y $ is the *codomain*.

It's important to note that not every element in the codomain needs to have a corresponding element in the domain. The elements of the codomain are simply the possible values that the function could output.

### Calculating the Domain of a Function

The domain of a function $f(x)$ is the set of all possible input values (values of $x$) for which the function is defined.

**Example**
>Consider the function $\sqrt{x^2 - 4}$
>
> Its natural domain is the set of $x$ values such that $x \leq -2 \lor x \geq 2$ 

## Explicit and Implicit Forms

- **Explicit Form:** In an explicit form, a function is defined by an algebraic expression that directly relates the input to the output. For example, the function $ f(x) = 2x + 3 $ is in explicit form. You can easily see how the input ($ x $) is transformed into the output.

- **Implicit Form:** In an implicit form, the relationship between the input and output may not be immediately apparent. Equations involving both the input and output variables can define functions implicitly. For example, the equation $ x^2 + y^2 = 1 $ implicitly defines a function because for each $ x $, there are two possible $ y $ values (considering both positive and negative square roots). Implicitly defined functions are often represented by equations involving both variables.

Understanding these concepts is crucial in various mathematical applications, including calculus, linear algebra, and real-world problem-solving. Functions provide a way to model and understand relationships between quantities and are fundamental in many areas of mathematics and science.


## The Graph of a Function
The graph of a function $f: X → Y$ is the subset of the Cartesian product $X × Y$ consisting of all pairs of the form $(x, f(x))$, that is, all and only the pairs ($x, y) ∈ X × Y$ that satisfy the equation $y = f(x)$.

Let's consider the function $f(x) = \sqrt{x^2 - 4}$:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):

    if -2 <= x <= 2:
        return np.nan
    
    return (x**2 - 4)**(1/2)

# Generate x values
x_values = np.linspace(-10, 10, 10000)

# Generate y values
y_values = np.vectorize(f)(x_values)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$f(x) = \sqrt{x^2 - 4}$', color='#5e7aad')

# Set the x and y limits
plt.xlim(-10, 10)
plt.ylim(-1, 10)

# Highlight critical points
plt.scatter([-2, 2], [0, 0], color='#ad6c6c', marker='o', label='Critical Points')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of the Function')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Show the plot
plt.show()
```

## Function Composition
Functions can be 'composed': if $f: X → Y$ and $g: Y → Z$ are functions, it makes sense to consider the composite function $g ◦ f: X → Z$, defined by $g ◦ f(x) = g(f(x))$ for every $x ∈ X$. Naturally, for the composition to make sense, the codomain of $f$ must be contained in the domain of $g$.

Now, let's consider an example. Let's take the function:

$$\Large f(x) = x^2$$
$$\Large f : \mathbb{R} \rightarrow \mathbb{R}^+$$

and the function

$$\Large g(x) = x + 2$$
$$\Large g : \mathbb{R} \rightarrow \mathbb{R}$$

Since $Im(f) \subset Dom(g)$ we can compose $f$ and $g$ to obtain a new function!

```python
import plotly.graph_objects as go
import numpy as np

# Create an array of x values from -10 to 10 to plot
x = np.linspace(-50, 50, 1000)

# Get the corresponding y values from the function f(x) = x^2
y1 = x**2

# Get the corresponding y values from the function f(x) = x + 2
y2 = x + 2

# Get the corresponding y values for the composition f(g(x))
y3 = y2**2

# Create a Plotly figure with three traces
fig = go.Figure()

# Add f(x) = x^2 to the plot with a specific color and a shaded area under the curve
fig.add_trace(go.Scatter(x=x, y=y1, name="f(x)=x^2", line=dict(color="blue")))

# Add f(x) = x + 2 to the plot with a different color and a shaded area under the curve
fig.add_trace(go.Scatter(x=x, y=y2, name="g(x)=x+2", line=dict(color="red")))

# Add the composition f(f(x)) to the plot with another color and a shaded area under the curve
fig.add_trace(go.Scatter(x=x, y=y3, name="h(x)", line=dict(color="green")))

# Add a vertical line at x=0
fig.add_shape(
    dict(type="line", x0=0, x1=0, y0=-50, y1=max(y1), line=dict(color="white"))
)

# Update title and axis labels
fig.update_layout(
    title="Graph of functions",
    xaxis_title="x",
    yaxis_title="y",
    width=1200,
    height=600,
    legend=dict(x=0.9, y=1),
    template='plotly_dark'
)

# Set graph layout options
fig.update_xaxes(range=[-20, 20])
fig.update_yaxes(range=[-50, 200])

# Show the Plotly figure
fig.show("png")
```

### Injectivity
A function $ f: X \rightarrow Y $ is said to be injective if distinct points are associated with distinct images, meaning:

$$ \large \forall x_0, x_1 \in X : f(x_0) = f(x_1) \Rightarrow x_0 = x_1 $$
$$ \text{or} $$
$$ \large \forall x_0, x_1 \in X : x_0 \neq x_1 \Rightarrow f(x_0) \neq f(x_1) $$

### Surjectivity
A function $ f: X \rightarrow Y $ is said to be surjective if $ f(X) = Y $, meaning that every $ y \in Y $ is the image of at least one $ x \in X $.

$$ \large \forall y \in Y \ \ \exists x \in X : f(x) = y $$

### Bijectivity
A function is called bijective, or invertible, or biunivocal, if it is both injective and surjective. In this case, for every $ y \in Y $, there is a unique $ x \in X $ such that $ f(x) = y $. In this scenario, the inverse function $ f^{-1} $ is defined (read as "f inverse"); $ f^{-1} $ is defined on $ Y $, taking values in $ X $, and associates each $ y \in Y $ with the unique $ x $ for which $ f(x) = y $. It is then said that $ f $ defines a biunivocal correspondence between the sets $ X $ and $ Y $. In particular, if $ f $ is bijective, it holds that $ f^{-1}(f(x)) = x $ for every $ x \in X $ and also $ f(f^{-1}(y)) = y $ for every $ y \in Y $: in other words, $ f^{-1} \circ f = I_X $, $ f \circ f^{-1} = I_Y $, where $ I_X $ and $ I_Y $ represent the identity functions on $ X $ and $ Y $, defined by $ I_X(x) = x $ for every $ x \in X $ and $ I_Y(y) = y $ for every $ y \in Y $.


Notably, the function $ f(x) = x^2 $ is not injective, as $ f(x) = f(x') \nRightarrow x = x' $. For example, $ \exists x, x' \in \mathbb{R} : f(x) = f(x') \land x \neq x' $, such as for $ x = 3 $ and $ x' = -3 $ where $ f(3) = f(-3) $.

However, it is observed that the function $ f(x) = x^2 $ is surjective, as $ \forall x \in \mathbb{R} \ \ \exists y \in [0, +\infty) : f(x) = y $.

Additionally, it is noted that it is not invertible since it is not bijective.

If $ f: X \rightarrow X $ is an invertible function, $ f^{-1}: X \rightarrow X $ is its inverse function. The equations $ y = f(x) $ and $ x = f^{-1}(y) $ are equivalent and both describe the graph of $ f $. Conversely, by interchanging the variables $ x $ and $ y $ (i.e., reflecting across the line $ y = x $ in the Cartesian plane $X \times X$), the second equation becomes $ y = f^{-1}(x) $, describing the graph of $ f^{-1} $, which is the symmetric image of the graph of $ f $ with respect to the bisector of the first quadrant.

It's worth noting that it is always possible to consider a given function $ f: X \rightarrow Y $ as surjective by regarding it as a function from $ X $ to its image $ f(X) $. The challenge lies in characterizing the subset $ f(X) $ of $ Y $, which can be difficult and sometimes impossible in concrete cases.
