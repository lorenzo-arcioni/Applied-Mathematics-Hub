# The Limit

The concept of **limit** is fundamental in differential calculus. It represents a way to describe the behavior of a function as the independent variable approaches a certain value. In other words, the limit tells us what happens to a function as its independent variable gets closer and closer to a particular point, but never actually reaches it.

## Notation

The standard notation for representing the limit of a function is as follows:

If we have a function $f(x)$ and want to calculate the limit of $f(x)$ as $x$ approaches a value $a$, we write:

$$
\Large\lim_{x \to a} f(x)
$$

In this notation, $\lim$ represents the limit, $x$ is the independent variable, and $a$ is the value toward which $x$ is approaching.

**Example:**
Consider the following function:

$$\Large f(x) = x^2 + x$$

```python
%matplotlib inline
from matplotlib import pyplot as plt

# Here's the function
def f(x):
    return x**2 + x

plt.figure(figsize=(12, 8))
# Create an array of x values from 0 to 10 to plot
x = list(range(0, 11))

# Get the corresponding y values from the function
y = [f(i) for i in x] 

# Set up the graph
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot the function
plt.plot(x,y, color='lightgrey', marker='o', markeredgecolor='green', markerfacecolor='green')

plt.show()
```

Calculate the limit:

$$\Large\lim_{x \to 5} \ x^2 + x$$

```python
%matplotlib inline

# Here's the function
def f(x):
    return x**2 + x

from matplotlib import pyplot as plt
plt.figure(figsize=(12, 8))
# Create an array of x values from 0 to 10 to plot
x = list(range(0,5))
x.append(4.25)
x.append(4.5)
x.append(4.75)
x.append(5)
x.append(5.25)
x.append(5.5)
x.append(5.75)
x = x + list(range(6,11))

# Get the corresponding y values from the function
y = [f(i) for i in x] 

# Set up the graph
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot the function
plt.plot(x,y, color='lightgrey', marker='o', markeredgecolor='green', markerfacecolor='green')

zx = 5
zy = f(zx)
plt.plot(zx, zy, color='red', marker='o', markersize=10)
plt.annotate('x=' + str(zx),(zx, zy), xytext=(zx - 0.5, zy + 5))

# Plot f(x) when x = 5.1
posx = 5.25
posy = f(posx)
plt.plot(posx, posy, color='blue', marker='<', markersize=10)
plt.annotate('x=' + str(posx),(posx, posy), xytext=(posx + 0.5, posy - 1))

# Plot f(x) when x = 4.9
negx = 4.75
negy = f(negx)
plt.plot(negx, negy, color='orange', marker='>', markersize=10)
plt.annotate('x=' + str(negx),(negx, negy), xytext=(negx - 1.5, negy - 1))

plt.show()
```

You can visualize the point where ***x*** is 5, and you may notice points on the graph that seem to be very close to this point (at *x=4.75* and *x=5.25*). However, if we were to zoom in, we would see that there are still empty spaces that could be filled by values of ***x*** even closer to 5, for example, 4.9 and 5.1, or 4.999 and 5.001. If we could zoom in infinitely, no matter how close the value you use (e.g., 4.999999999999), there is always a fractionally closer value (e.g., 4.9999999999999).

So, what we can say is that there is a hypothetical number that is as close as possible to our desired value of *x* without actually being *x*, but we cannot express it as a real number. Instead, we symbolically express it as a *limit*, like this:

\begin{equation}\Large\lim_{x \to 5} f(x) = 25\end{equation}

This is interpreted as *the limit of the function f(x) as *x* approaches 5*.

### Applications

The concept of a limit is widely used in differential calculus to define the derivative of a function and to solve problems related to continuity and the behavior of functions at critical points. It is a fundamental basis for understanding change and the rate of change of functions, which are central concepts in calculus.

In summary, the limit is a crucial mathematical tool that allows us to analyze the behavior of functions as their independent variable approaches a specific point, even if it never reaches it.

#### Limits and Continuity
The function ***f(x)*** is *continuous* for all real values of ***x***. In simple terms, this means that you can draw the line created by the function without lifting the pen (we will examine a more formal definition later in this course).

However, this is not necessarily true for all functions. Consider the function ***g(x)*** below:

\begin{equation}g(x) = -\left(\frac{12}{2x}\right)^{2}\end{equation}

This function is a bit more complex than the previous one, but the key thing to note is that it involves a division by *2x*. Now, ask yourself: what would happen if you applied this function to a value of *x* equal to **0**?

Well, anything divided by 0 is *undefined*. So the *domain* of this function does not include 0; in other words, the function is defined when *x* is any real number such that *x is not equal to 0*. The function should then be written as:

\begin{equation}g(x) = -\left(\frac{12}{2x}\right)^{2},\;\; x \ne 0\end{equation}

```python
%matplotlib inline

# Define function g
def g(x):
    if x != 0:
        return -(12/(2*x))**2
    
# Plot output from function g
from matplotlib import pyplot as plt

# Create an array of x values
x = range(-20, 21)


# Get the corresponding y values from the function
y = [g(a) for a in x]

# Set up the graph
plt.figure(figsize=(12, 8))
plt.xlabel('x')
plt.ylabel('g(x)')

plt.xlim(-20, 20)
plt.ylim(-20, 10)

plt.grid(True)

# Plot x against g(x)
plt.plot(x,y, color='#366e42')

# plot a circle at the gap (or close enough anyway!)
xy = (-0.555, -20)
plt.annotate('O',xy=xy,fontsize=24,color='#425987')

plt.show()
```

Carefully observe the graph and notice the gap in the line at *x* = 0. This indicates that the function is not defined at this point. The *domain* of the function (i.e., the set of possible input values) does not include 0, and its *codomain* (the set of possible output values) does not include a value for x=0.

It is a *non-continuous* function— in other words, it has at least one gap when plotted (so you couldn't trace it by hand without lifting the pen). Specifically, the function is not continuous at x=0.

By convention, when plotting a non-continuous function, the points that form a continuous line (or *interval*) are shown as a line, while the end of each interval where there is a discontinuity is represented by a circle, which is filled if the value at that point is included in the line and empty if the value is not included in the line.

In this case, the function produces two intervals with a gap between them where the function is not defined, so we can represent the point of discontinuity as an unfilled circle.

#### Intuitive Interpretation

The main idea of the limit is to determine the value that a function approaches as the independent variable gets closer and closer to a specific point. This concept is useful because some functions may exhibit complex behaviors or discontinuities at specific points, and the limit allows us to better understand how the function behaves near those points.

For example, let's consider the function $f(x) = \frac{1}{x}$. As $x$ approaches 0 from the right ($0^+$), this function becomes very large (positive); whereas, as $x$ approaches 0 from the left ($0^-$), this function becomes very small (negative), and it appears to diverge. However, using the concept of a limit, we can say that:

$$
\Large\lim_{x \to 0^+} \ \frac{1}{x} \ = +\infty 
$$
$$
\Large\lim_{x \to 0^-} \ \frac{1}{x} \ = -\infty \\
$$
$$
\Large\lim_{x \to 0} \ \frac{1}{x} \ = Does \ not \ exist
$$

This means that the function tends to positive infinity as $x$ approaches 0 from the right, while it tends to negative infinity as $x$ approaches 0 from the left. We can state that the general limit does not exist since the limits from the right and left are different.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Creazione di un array di valori x
x_values = np.linspace(-2, 2, 30000)  # Consideriamo un intervallo simmetrico intorno a 0

# Calcolo della funzione per i valori di x
y_values = 1 / x_values

# Creazione del grafico Seaborn
sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))

# Funzione originale
plt.scatter(x=x_values, y=y_values, label='f(x) = 1/x', color='#4b5b7a', linewidth=2, s=3)

# Linea verticale a x = 0
plt.axvline(x=0, color='black', linestyle='--')

# Etichette e legenda
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Aggiungi annotazioni per alcuni punti specifici
annotations = [
    [(-1, -7.5), 'Limit from left -->'],  # Esempio di annotazione per il punto (-1, -1)
    [(0.5, 7.5), '<-- Limit from right']   # Esempio di annotazione per il punto (0.5, 2)
]

for annotation in annotations:
    plt.annotate(annotation[1], xy=annotation[0], fontsize=12, color='#664228')


# Mostra il grafico
plt.title('Graph of f(x) = 1/x')
plt.xlim(-2, 2)
plt.ylim(-10, 10)
plt.grid(True)

plt.show()
```

## Rules for Operations on Limits
When working with functions and limits, you may want to combine limits using arithmetic operations. There are some intuitive rules to do so.

Let's define two simple functions, ***j***:

\begin{equation}
j(x) = 2x - 2
\end{equation}

and ***l***:

\begin{equation}
l(x) = -2x + 4
\end{equation}

Run the cell below to plot these functions:

```python
%matplotlib inline

# Define function j
def j(x):
    return x * 2 - 2

# Define function l
def l(x):
    return -x * 2 + 4


# Plot output from functions j and l
from matplotlib import pyplot as plt

# Create an array of x values
x = range(-10, 11)

# Get the corresponding y values from the functions
jy = [j(i) for i in x]
ly = [l(i) for i in x]

# Set up the graph
plt.xlabel('x')
plt.xticks(range(-10,11, 1))
plt.ylabel('y')
plt.yticks(range(-30,30, 2))
plt.grid()

# Plot x against j(x)
plt.plot(x,jy, color='green', label='j(x)')

# Plot x against l(x)
plt.plot(x,ly, color='#4b5b7a', label='l(x)')

plt.legend()

plt.show()
```

### Sum (and Subtraction) of Limits
Let's start with the rule for sum:

\begin{equation}
\lim_{x \to a} (j(x) + l(x)) = \lim_{x \to a} j(x) + \lim_{x \to a} l(x)
\end{equation}

What we are saying here is that the limit of *j(x)* + *l(x)* as *x* approaches *a* is the same as the limit of *j(x)* as *x* approaches *a*, plus the limit of *l(x)* as *x* approaches *a*.

Looking at the graph of our functions ***j*** and ***l***, let's apply this rule at a value of *a* equal to **8**.

Visually inspecting the graph, you can see that as *x* approaches 8 from both directions, *j(x)* approaches 14. So:

\begin{equation}
\lim_{x \to 8} j(x) = 14
\end{equation}

Similarly, as *x* approaches 8 from both directions, *l(x)* approaches -12. So:

\begin{equation}
\lim_{x \to 8} l(x) = -12
\end{equation}

So, based on the sum rule:

\begin{equation}
\lim_{x \to 8} (j(x) + l(x)) = 14 + (-12) = 2
\end{equation}

### Multiplication of Limits
Here's the rule for multiplication:

\begin{equation}
\lim_{x \to a} (j(x) \cdot l(x)) = \lim_{x \to a} j(x) \cdot \lim_{x \to a} l(x)
\end{equation}

Again, you can apply this rule to the limits as *x* approaches a value *a* of 8 that we identified earlier:

\begin{equation}
\lim_{x \to 8} (j(x) \cdot l(x)) = 14 \cdot -12 = -168
\end{equation}

This rule also applies when multiplying a limit by a constant:

\begin{equation}
\lim_{x \to a} c \cdot l(x) = c \cdot \lim_{x \to a} l(x)
\end{equation}

So, for a value *a* of 8 and a constant *c* of 3, this is equivalent to:

\begin{equation}
\lim_{x \to 8} 3 \cdot l(x) = 3 \cdot -12 = -36
\end{equation}

### Division of Limits
For division, assuming the limit of *l(x)* as *x* approaches *a* is not 0:

\begin{equation}
\lim_{x \to a} \frac{j(x)}{l(x)} = \frac{\lim_{x \to a} j(x)}{\lim_{x \to a} l(x)}
\end{equation}

So, based on the limits for *j(x)* and *l(x)* as *x* approaches 8:

\begin{equation}
\lim_{x \to 8} \frac{j(x)}{l(x)} = \frac{14}{-12}= \frac{7}{-6}
\end{equation}

### Limits of Exponentials and Roots

Assuming *n* is an integer:

\begin{equation}
\lim_{x \to a} j(x)^{n} = \Big(\lim_{x \to a} j(x)\Big)^{n}
\end{equation}

So, for example:

\begin{equation}
\lim_{x \to 8} j(x)^{2} = \Big(\lim_{x \to 8} j(x)\Big)^{2} = 14^{2} = 196
\end{equation}

For roots, again assuming *n* is an integer:

\begin{equation}
\lim_{x \to a} \sqrt[n]{j(x)} = \sqrt[n]{\lim_{x \to a} j(x)}
\end{equation}

So:

\begin{equation}
\lim_{x \to 8} \sqrt[2]{j(x)} = \sqrt[2]{\lim_{x \to 8} j(x)} \approx 3.74
\end{equation}

## Derivazione e Derivate
Finora in questo corso, hai imparato come valutare i limiti per punti su una retta. Ora stai per ampliare questa conoscenza e esaminare una tecnica di calcolo chiamata *derivazione*. Nella derivazione, utilizziamo la nostra conoscenza dei limiti per calcolare la *derivata* di una funzione al fine di determinare il tasso di cambiamento in un punto individuale su una retta.

Ricordiamoci del problema che stiamo cercando di risolvere, ecco una funzione:

\begin{equation}
f(x) = x^{2} + x
\end{equation}

Possiamo visualizzare una parte di questa funzione utilizzando il seguente codice Python:

```python
%matplotlib inline

# Here's the function
def f(x):
    return x**2 + x

from matplotlib import pyplot as plt

# Create an array of x values from 0 to 10 to plot
x = list(range(0, 11))

# Use the function to get the y values
y = [f(i) for i in x]

# Set up the graph
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot the function
plt.plot(x,y, color='green')

plt.show()
```

We know that we can calculate the average rate of change for a given interval on the line by determining the slope for a secant line connecting two points on the line. For example, we can calculate the average change for the interval between x=4 and x=6 by dividing the change (or *delta*, denoted as &Delta;) in the value of *f(x)* by the change in the value of *x*:

\begin{equation}
m = \frac{\Delta{f(x)}}{\Delta{x}}
\end{equation}

The delta for *f(x)* is calculated by subtracting the *f(x)* values of our points, and the delta for *x* is calculated by subtracting the *x* values of our points; as follows:

\begin{equation}
m = \frac{f(x_{2}) - f(x_{1})}{x_{2} - x_{1}}
\end{equation}

So, for the interval between x=4 and x=6, we would have:

\begin{equation}
m = \frac{f(6) - f(4)}{6 - 4} =  \frac{42 - 20}{6 - 4} = 11
\end{equation}

We can calculate and plot this using the following Python code:

```python
%matplotlib inline

def f(x):
    return x**2 + x

from matplotlib import pyplot as plt

# Create an array of x values from 0 to 10 to plot
x = list(range(0, 11))

# Use the function to get the y values
y = [f(i) for i in x]

# Set the a values
x1 = 4
x2 = 6

# Get the corresponding f(x) values 
y1 = f(x1)
y2 = f(x2)

# Calculate the slope by dividing the deltas
a = (y2 - y1)/(x2 - x1)

# Create an array of x values for the secant line
sx = [x1,x2]

# Use the function to get the y values
sy = [f(i) for i in sx]

# Set up the graph
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()

# Plot the function
plt.plot(x,y, color='green')

# Plot the interval points
plt.scatter([x1,x2],[y1,y2], c='red')

# Plot the secant line
plt.plot(sx,sy, color='magenta')

# Display the calculated average rate of change
plt.annotate('Average change =' + str(a),(x2, (y2+y1)/2))

plt.show()
```

The average rate of change for the interval between x=4 and x=6 is <sup>11</sup>/<sub>1</sub> (or simply 11), which means that for every **1** added to *x*, *f(x)* increases by **11**. In other words, if x represents time in seconds and f(x) represents distance in meters, the average rate of change for distance over time (i.e., *velocity*) for the interval from 4 to 6 seconds is 11 meters per second.
