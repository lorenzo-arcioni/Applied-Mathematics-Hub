# Probabilità 1

La probabilità è una branca della matematica e della statistica che si occupa di quantificare l'incertezza e la casualità in situazioni dove gli esiti futuri sono incerti. In termini semplici, la probabilità fornisce uno strumento per misurare quanto sia probabile che un certo evento si verifichi. Si basa su un insieme di principi e regole che consentono di assegnare numeri alle diverse possibilità di risultati. Lo spazio di probabilità è dunque formato da tre elementi:
  - Lo spazio campione $\mathcal{\Omega}$
  - Lo spazio degli eventi $\mathcal{F}$
  - Legge di probabilità $\mathbb{P}$


$$\Large (\Omega, \ \mathcal{F}, \ \mathbb{P})$$

### Spazio campione

Lo **spazio campione** rappresenta l'insieme di tutti i possibili risultati di un esperimento casuale. Ad esempio, se lanci un dado equo a sei facce, lo spazio campione sarebbe {1, 2, 3, 4, 5, 6}, poiché questi sono i sei risultati possibili.

Gli **eventi**, denotati con il simbolo $E_i$, sono sottoinsiemi dell'**spazio campione** che rappresentano gli esiti che ci interessano. Ad esempio, se vogliamo calcolare la probabilità di ottenere un "6" quando lanciamo il dado, l'evento corrispondente sarebbe {6}.

### Spazio degli eventi

L'insieme di tutti i possibili eventi che possono verificarsi, indicato con il simbolo $\mathcal{F}$.

Quindi

$$\Large E_i \in \mathcal{F}$$

$$ \Large \mathcal{F} = \mathcal{P}(\Omega) \rightarrow insieme \ delle \ parti \ di \ \Omega$$

### Legge di probabilità

Una legge che associa ad ogni evento la probabilità che esso si verifichi.

$$\Large \mathbb{P}: \mathcal{F} \ \rightarrow \ [0, \ 1] $$

### Variabili aleatorie

Una **variabile aleatoria** è una funzione che assegna un valore numerico a ciascun elemento dello spazio campione. In altre parole, è una mappatura degli esiti possibili a numeri reali. Ci sono due tipi principali di variabili aleatorie:

1. **Variabili aleatorie discrete**: Queste variabili assumono solo un numero finito o numerabile di valori. Ad esempio, la variabile aleatoria che rappresenta il risultato di un lancio di dado è discreta poiché può assumere solo i valori 1, 2, 3, 4, 5 o 6.

2. **Variabili aleatorie continue**: Queste variabili possono assumere un numero infinito non numerabile di valori all'interno di un intervallo. Ad esempio, l'altezza di una persona è una variabile aleatoria continua poiché può assumere un numero infinito di valori all'interno di un intervallo di altezza.

Quindi la variabile aleatoria $X$, ad esempio, è così definita:

$$\Large
X: \Omega \rightarrow c, \ c \in \mathbb{R}
$$

## Esempi

### Spazio Campione ed Eventi

Supponiamo di lanciare un dado equo a sei facce. In questo caso:

- Lo spazio campione Ω è l'insieme dei risultati possibili: Ω = {1, 2, 3, 4, 5, 6}.

- Gli eventi possono essere rappresentati come sottoinsiemi di Ω:
  - L'evento A di ottenere un numero pari: A = {2, 4, 6}.
  - L'evento B di ottenere un numero maggiore di 4: B = {5, 6}.

### Variabili Aleatorie

Supponiamo di definire una variabile aleatoria $X$ che rappresenta il risultato di un lancio di dado. La funzione che associa il risultato del dado a un numero reale è la seguente:

- $X$(1) = 1
- $X$(2) = 2
- $X$(3) = 3
- $X$(4) = 4
- $X$(5) = 5
- $X$(6) = 6

Questa è una variabile aleatoria discreta perché può assumere solo un numero finito di valori.

### Legge di Probabilità

La Legge di probabilità di $X$ può essere rappresentata come segue:

$$ \Large \mathbb{P}(X = x) =
\begin{cases}
\frac{1}{6} & \text{se } x \in \{1, 2, 3, 4, 5, 6\} \\
0 & \text{altrimenti}
\end{cases}$$

Questo significa che la probabilità di ottenere ciascun valore da 1 a 6 è uguale a 1/6.

### Funzione di Densità di Probabilità
La PDF descrive come le probabilità sono distribuite su un insieme infinito di valori possibili.
Supponiamo di avere una variabile aleatoria continua $Y$ che rappresenta l'altezza di una persona in centimetri. La sua funzione di densità di probabilità (pdf) potrebbe essere modellata da una distribuzione normale (gaussiana) con media μ e deviazione standard σ. La pdf di $Y$ che assume il valore $y \in \mathbb{R}$ sarebbe:

$$
\Large f(y) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y - \mu)^2}{2\sigma^2}}
$$

Dove:
  - $\mu$ è il valore medio della distribuzione.
  - $\sigma$ è la devizione standard della distribuzione.

```python
# Esempio di spazio campione e eventi

# Importa librerie necessarie
import random

# Definizione dello spazio campione per il lancio di un dado a sei facce
spazio_campione = [1, 2, 3, 4, 5, 6]

# Esempio di un evento: Ottenere un numero pari
evento_A = [x for x in spazio_campione if x % 2 == 0]
print("Spazio Campione:", spazio_campione)
print("Evento A (numeri pari):", evento_A)

# Esempio di un evento: Ottenere un numero maggiore di 4
evento_B = [x for x in spazio_campione if x > 4]
print("Evento B (numeri maggiori di 4):", evento_B)


# Esempio di variabile aleatoria

# Definizione di una variabile aleatoria per il lancio di un dado
def variabile_aleatoria_dado():
    return random.choice(spazio_campione)

# Esempio di utilizzo della variabile aleatoria
risultato_lancio = variabile_aleatoria_dado()
print("Risultato del lancio del dado:", risultato_lancio)


# Esempio di distribuzione di probabilità

# Calcolo della probabilità di ottenere un "3" nel lancio del dado
probabilita_3 = 1 / len(spazio_campione)
print("Probabilità di ottenere un 3:", round(probabilita_3, 4), "ossia", round(probabilita_3 * 100, 2), "%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Media e deviazione standard della distribuzione normale
media_altezza = 170
deviazione_standard = 10

# Generazione di dati casuali basati su una distribuzione normale
dati_altezza = np.random.normal(media_altezza, deviazione_standard, 10000)

# Creazione di un istogramma per rappresentare la pdf
plt.figure(figsize=(10, 6))  # Imposta la dimensione dell'immagine
plt.hist(dati_altezza, bins=30, density=True, alpha=0.5, color='b', label='Dati Altezza')

# Calcolo dei valori della pdf della distribuzione normale
x = np.linspace(media_altezza - 4 * deviazione_standard, media_altezza + 4 * deviazione_standard, 100)
pdf = norm.pdf(x, media_altezza, deviazione_standard)

# Sovrapposizione della curva della pdf
plt.plot(x, pdf, 'r', label='PDF della Distribuzione Normale')

plt.title("Funzione di Densità di Probabilità (pdf) dell'Altezza")
plt.xlabel("Altezza (cm)")
plt.ylabel("Densità di Probabilità")
plt.legend(loc='upper left')  # Sposta la legenda in alto a sinistra
plt.show()
```

#### Densità di Probabilità: Un Esempio Intuitivo

La **densità di probabilità** è un concetto fondamentale nella teoria delle probabilità che misura quanto sia probabile che una variabile casuale cada in un certo intervallo. Possiamo comprenderlo intuitivamente pensando a un esempio con l'altezza delle persone.

Supponiamo di avere una popolazione di persone e di misurare le loro altezze. Vogliamo capire quanto è probabile trovare persone in diverse fasce di altezza. La densità di probabilità ci aiuta in questo modo:

- Se la densità di probabilità dell'altezza è **alta** in un intervallo, significa che è probabile che molte persone abbiano altezze in quella fascia. Troveremo molte persone con altezze in quell'intervallo.

- Se la densità di probabilità dell'altezza è **bassa** in un intervallo, significa che è meno probabile trovare persone con altezze in quella fascia. Troveremo meno persone con altezze in quell'intervallo.

In breve, la densità di probabilità ci fornisce una misura della "concentrazione" della probabilità in diverse parti del dominio della variabile casuale. Le aree sotto la curva della densità di probabilità che sono più alte rappresentano gli intervalli in cui la variabile casuale ha una maggiore probabilità di assumere valori.

È importante notare che l'area totale sotto la curva della densità di probabilità deve essere **uguale a 1**, poiché la somma di tutte le probabilità possibili deve essere 1 (cioè la somma di tutte le possibilità è sicuramente 1).


### Densità di Probabilità e Probabilità in una Distribuzione Normale

La **densità di probabilità** (pdf) in una distribuzione normale (gaussiana) è definita dalla seguente formula:

$$
\large f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Dove:
- $f(x)$ è la densità di probabilità di $x$.
- $x$ è il valore specifico di $X$.
- $\mu$ è la media della distribuzione.
- $\sigma$ è la deviazione standard della distribuzione.

Per calcolare la **probabilità di trovare valori in un intervallo specifico** $[a, b]$ all'interno di una distribuzione normale, utilizziamo l'integrale definito:

$$
\mathbb{P}(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx
$$

Dove:
- $X$ è la variabile casuale.
- $a$ e $b$ sono i limiti dell'intervallo di interesse.
- $\int$ rappresenta l'integrale definito su quell'intervallo.

Ad esempio, se vogliamo calcolare la probabilità di trovare altezze tra 160 cm e 180 cm in una distribuzione normale delle altezze con una media $\mu$ e una deviazione standard $\sigma$, possiamo calcolare l'integrale definito della pdf tra questi limiti di intervallo.

Nel nostro caso, $\mu$ rappresenta la media delle altezze e $\sigma$ la deviazione standard. Supponiamo di avere $\mu = 170$ cm e $\sigma = 10$ cm.

L'intervallo di interesse è $[160, 180]$.

Ora calcoliamo l'integrale definito per trovare la probabilità:

$$
\mathbb{P}(160 \leq X \leq 180) = \int_{160}^{180} \frac{1}{10 \sqrt{2\pi}} e^{-\frac{(x - 170)^2}{2 \cdot 10^2}} \, dx
$$

### Dimostriamo che l'integrale della PDF tra $+\infty$ e $-\infty$ è uguale a 1

La PDF data è:
$$\Large f(y) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y - \mu)^2}{2\sigma^2}}$$

Ora, calcoliamo l'integrale della PDF da meno infinito a più infinito:

$$
\begin{aligned}
\int_{-\infty}^{\infty} f(y) dy &= \int_{-\infty}^{\infty} \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(y - \mu)^2}{2\sigma^2}} dy = \frac{1}{\sigma \sqrt{2\pi}} \int_{-\infty}^{\infty} e^{-\frac{(y - \mu)^2}{2\sigma^2}} dy = \frac{1}{\sigma \sqrt{2\pi}} \sqrt{2\pi}\sigma = 1\\
\end{aligned}
$$

Quindi, l'integrale della PDF da meno infinito a più infinito è uguale a 1, dimostrando che la PDF è normalizzata correttamente.

```python
import numpy as np
from scipy.stats import norm
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Parametri della distribuzione normale
media = 170
deviazione_standard = 10

# Definizione della pdf (probability density function) della distribuzione normale
def pdf(x):
    return norm.pdf(x, loc=media, scale=deviazione_standard)

# Creazione di un array di valori x per la curva
x = np.linspace(media - 5 * deviazione_standard, media + 5 * deviazione_standard, 1000)

# Calcolo dei valori della pdf
pdf_values = pdf(x)

# Creazione del grafico della pdf
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_values, 'b', label='Densità di Probabilità (pdf)')
plt.title("Distribuzione Normale delle Altezze")
plt.xlabel("Altezza (cm)")
plt.ylabel("Densità di Probabilità")

colors = ['r', 'b', 'g', 'y']

# Aggiunta delle linee verticali per gli intervalli
intervalli = [(160, 180), (150, 190), (135, 200), (120, 220)]
for intervallo,colore in zip(intervalli, colors):
    probabilita, _ = quad(pdf, intervallo[0], intervallo[1])
    plt.axvline(intervallo[0], color=colore, linestyle='--', label=f'{probabilita*100:.2f}%')
    plt.axvline(intervallo[1], color=colore, linestyle='--')

# Aggiunta di una legenda
plt.legend()
plt.grid(True)
plt.show()
```

## Indipendenza e Dipendenza tra Eventi

Nella teoria delle probabilità, è importante comprendere la relazione tra **indipendenza e dipendenza tra eventi**. Questo concetto è cruciale per valutare come gli eventi sono correlati tra loro e influenzano la probabilità complessiva di un insieme di eventi.

### Eventi Indipendenti

Due eventi sono considerati **indipendenti** se l'occorrenza o la non-occorrenza di uno di essi **non influisce** sull'occorrenza dell'altro. In altre parole, la probabilità congiunta degli eventi indipendenti è il prodotto delle probabilità di ciascun evento separatamente.

La definizione matematica di eventi indipendenti è:

$$\large\mathbb{P}(A \cap B) = \mathbb{P}(A) \cdot \mathbb{P}(B)$$

**Esempio**: Supponiamo di lanciare due dadi equi a sei facce. Gli eventi "ottenere un 3 al primo lancio" e "ottenere un 4 al secondo lancio" sono indipendenti.

```python
# Probabilità di ottenere un 3 al primo lancio
P_A = 1 / 6

# Probabilità di ottenere un 4 al secondo lancio
P_B = 1 / 6

# Evento A x B: Lancio un dado e esce 3 (1/6) lancio l'altro ed esce 4 (1/6)
# Calcolo della probabilità moltiplicata di A e B (eventi indipendenti)
A_moltiplicato_B = P_A * P_B

print("La probabilità moltiplicata di A e B (eventi indipendenti) è:", A_moltiplicato_B)

# Calcolo della probabilità congiunta di A e B (eventi indipendenti)
# Evento A: Ottengo 3 con il primo dado -> ottengo 3 con un dado -> P(A) = 1/6
# Evento B: Ottengo 4 con il secondo dado -> ottengo 4 con un dado -> P(B) = 1/6
# Evento A ∩ B: Lancio un dado e esce 3 (1/6) lancio l'altro ed esce 4 (1/6)

A_intersezione_B = P_A * P_B

print("La probabilità intersezione di A e B (eventi indipendenti) è:", A_intersezione_B)
```

### Eventi Dipendenti

Gli eventi sono dipendenti quando l'occorrenza o la non-occorrenza di uno di essi influenza l'occorrenza dell'altro. In questo caso, le probabilità congiunte non possono essere calcolate semplicemente come il prodotto delle probabilità individuali.

$$\large\mathbb{P}(A \cap B) \neq \mathbb{P}(A) \cdot \mathbb{P}(B)$$

Un esempio comune di eventi dipendenti è il campionamento senza reinserimento da una popolazione. Ad esempio, se si estraggono due carte da un mazzo di carte senza reinserimento, la probabilità del secondo evento dipenderà dal risultato del primo.

```python
# Probabilità di estrarre un asso dal mazzo (primo estratto)
P_A = 4 / 52

# Probabilità di estrarre un asso dal mazzo (secondo estratto senza reinserimento)
P_B_dato_A = 3 / 51  # Poiché c'è un asso in meno nel mazzo

print("La probabilità di estrarre un asso dal mazzo completo:", P_A)
print("La probabilità di estrarre un asso dal mazzo (senza l'asso estratto):", P_B_dato_A)
```

## Teorema di probabilità assoluta

Il teorema di probabilità assoluta è un concetto fondamentale nella teoria delle probabilità. Per calcolare la probabilità di un evento A, quando ci sono molteplici cause o condizioni possibili (B1, B2, ..., Bn), possiamo utilizzare la seguente formula:

$$\Large
P(A) = \sum_{i=1}^{n} P(A | B_i) \cdot P(B_i)
$$

Dove:
- $P(A)$ è la probabilità dell'evento A.
- $P(A | B_i)$ è la probabilità condizionale di A dato B_i.
- $P(B_i)$ è la probabilità dell'evento B_i.

Questo teorema ci permette di considerare tutte le possibili cause (B1, B2, ..., Bn) e calcolare la probabilità di A tenendo conto delle probabilità condizionali e delle probabilità delle singole cause.


**Esempio di Calcolo della Probabilità di Ottenere un Punteggio Positivo in un Esame**

In questo esempio, dimostriamo come calcolare la probabilità di ottenere un punteggio positivo in un esame basata su due possibili condizioni: "Studiare" e "Non Studiare". Supponiamo che ci siano due condizioni possibili prima dell'esame, ovvero, uno studente può decidere di studiare o non studiare. Abbiamo le seguenti probabilità e probabilità condizionali:

- $P(\text{Positivo | Studiare})$: La probabilità di ottenere un punteggio positivo se si studia.
- $P(\text{Positivo | Non Studiare})$: La probabilità di ottenere un punteggio positivo se non si studia.
- $P(\text{Studiare})$: La probabilità di studiare prima dell'esame.
- $P(\text{Non Studiare})$: La probabilità di non studiare prima dell'esame.

Utilizzando il teorema di probabilità assoluta, calcoleremo la probabilità di ottenere un punteggio positivo in base alle probabilità delle condizioni e alle probabilità condizionali.

```python
# Probabilità condizionale P(Positivo | Studiare)
P_Positivo_dato_Studiare = 0.75  # Probabilità di ottenere un punteggio positivo se si studia

# Probabilità condizionale P(Positivo | Non Studiare)
P_Positivo_dato_Non_Studiare = 0.25  # Probabilità di ottenere un punteggio positivo se non si studia

# Probabilità marginale P(Studiare)
P_Studiare = 0.8  # Probabilità di studiare prima dell'esame

# Probabilità marginale P(Non Studiare)
P_Non_Studiare = 0.2  # Probabilità di non studiare prima dell'esame

# Calcolare la probabilità di ottenere un punteggio positivo
P_Positivo = (P_Positivo_dato_Studiare * P_Studiare) + (P_Positivo_dato_Non_Studiare * P_Non_Studiare)

print("La probabilità di ottenere un punteggio positivo è di:", round(P_Positivo*100, 2), "%")
```

## Teorema della probabilità totale

Il teorema della probabilità dell'unione, o totale, è una regola fondamentale della teoria della probabilità che stabilisce quanto segue:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

dove:
- $P(A \cup B)$ rappresenta la probabilità dell'unione degli eventi $A$ e $B$.
- $P(A)$ è la probabilità dell'evento $A$.
- $P(B)$ è la probabilità dell'evento $B$.
- $P(A \cap B)$ è la probabilità dell'intersezione degli eventi $A$ e $B$.

Questo teorema tiene conto del fatto che quando calcoliamo la probabilità dell'unione di due eventi, stiamo contando la probabilità di ciascun evento separatamente, ma dobbiamo sottrarre la probabilità dell'intersezione una volta, poiché altrimenti l'intersezione verrebbe conteggiata due volte.


```python
# Probabilità di ottenere un numero pari su un dado a sei facce
P_A = 3/6  # 1/2

# Probabilità di ottenere un numero maggiore di 3 su un dado a sei facce
P_B = 3/6  # 1/2

# Probabilità di ottenere un numero pari e maggiore di 3 contemporaneamente
P_A_intersezione_B = 2/6  # 1/3

# Calcolare la probabilità dell'unione di A e B utilizzando il teorema dell'unione
P_A_unione_B = P_A + P_B - P_A_intersezione_B

# Stampare il risultato
print("La probabilità dell'unione di A e B è:", round(P_A_unione_B*100, 2), "%")
```

### La probabilità condizionata
La **probabilità condizionata** è un concetto fondamentale della teoria delle probabilità e si riferisce alla probabilità di un evento $A$ che si verifica, dato che un certo evento $B$ si è già verificato o è noto. Si indica come $\mathbb{P}(A | B)$ e si legge come "la probabilità di A dato B".

La formula generale per calcolare la probabilità condizionata è data da:

$$\Large\mathbb{P}(A | B) = \frac{\mathbb{P}(A \cap B)}{\mathbb{P}(B)}$$

Dove:
- $\mathbb{P}(A | B)$ è la probabilità di A dato B.
- $\mathbb{P}(A \cap B)$ è la probabilità dell'intersezione di A e B.
- $\mathbb{P}(B)$ è la probabilità di B.

In parole semplici, la probabilità condizionata ci dice quanto è probabile che si verifichi l'evento A, sapendo che l'evento B si è già verificato. La probabilità condizionata tiene conto dell'informazione fornita da B e può essere diversa dalla probabilità di A senza alcuna informazione aggiuntiva (ossia, senza la conoscenza di B).

In generale abbiamo che:
  - $[ \ \mathbb{P}(X | Y) = \mathbb{P}(X) \land \mathbb{P}(Y | X) = \mathbb{P}(Y) \ ]\iff$ i due eventi sono indipendenti
  - $[ \ \mathbb{P}(X | Y) \neq \mathbb{P}(X) \lor \mathbb{P}(Y | X) \neq \mathbb{P}(Y) \ ]\iff$ i due eventi sono dipendenti

Notiamo che $\mathbb{P}(A \cap B)$ può essere scritto come:
  - $\mathbb{P}(A | B) \cdot \mathbb{P}(B)$ oppure
  - $\mathbb{P}(B | A) \cdot \mathbb{P}(A)$


## Teorema di Bayes
Il **Teorema di Bayes** è uno strumento fondamentale nella teoria delle probabilità e nella statistica che ci permette di aggiornare le nostre credenze o stime in base a nuove informazioni.
Il teorema di Bayes è espresso dalla seguente formula:

$$\Large
\mathbb{P}(A | B) = \frac{\mathbb{P}(B | A) \cdot \mathbb{P}(A)}{\mathbb{P}(B)} \  \ con \ \mathbb{P}(B) \neq 0
$$

Dove:
- $\mathbb{P}(A | B)$ rappresenta la probabilità dell'evento A dato che si è verificato l'evento B. Questo è chiamato la "probabilità condizionata" di A dato B.
- $\mathbb{P}(B | A)$ rappresenta la probabilità dell'evento B dato che si è verificato l'evento A.
- $\mathbb{P}(A)$ è la probabilità iniziale di A (prima di considerare B).
- $\mathbb{P}(B)$ è la probabilità iniziale di B (prima di considerare A).

Il teorema di Bayes è utile quando vogliamo calcolare la probabilità di un'ipotesi A alla luce di una nuova evidenza B. In altre parole, ci permette di aggiornare la nostra conoscenza o credenza su A in base a ciò che sappiamo su B.

Il teorema di Bayes è particolarmente potente quando si tratta di problemi di inferenza statistica, come il calcolo delle probabilità a posteriori o il riconoscimento di modelli in dati complessi.

### Esempio: Supponiamo di avere un test medico per una malattia rara. 

#### Probabilità iniziali
- $\mathbb{P}(\text{Malattia}) = 0.01$  # Probabilità di avere la malattia
- $\mathbb{P}(\text{Negativo | Malattia}) = 0.05$  # Probabilità di risultato negativo dato che si ha la malattia (falso negativo)
- $\mathbb{P}(\text{Negativo | No Malattia}) = 0.80$  # Probabilità di risultato negativo dato che non si ha la malattia (vero negativo)

#### Calcolo della probabilità a posteriori usando il Teorema di Bayes
 - $\mathbb{P}(\text{No Malattia}) = 1 - \mathbb{P}(\text{Malattia})$  # Probabilità di non avere la malattia
 - $\mathbb{P}(\text{Negativo}) = \mathbb{P}(\text{Negativo | Malattia}) \cdot \mathbb{P}(\text{Malattia}) + \mathbb{P}(\text{Negativo | No Malattia}) \cdot \mathbb{P}(\text{No Malattia})$
 - $\mathbb{P}(\text{Malattia | Negativo}) = \frac{(\mathbb{P}(\text{Negativo | Malattia}) \cdot \mathbb{P}(\text{Malattia}))}{\mathbb{P}(\text{Negativo})}$

```python
# Probabilità iniziali
p_malattia = 0.1  # Probabilità di avere la malattia
p_negativo_malattia = 0.05  # Probabilità di risultato negativo dato che si ha la malattia
p_negativo_no_malattia = 0.80  # Probabilità di risultato negativo dato che non si ha la malattia

# Calcolo della probabilità a posteriori usando il Teorema di Bayes
p_no_malattia = 1 - p_malattia  # Probabilità di non avere la malattia
p_negativo = (p_negativo_malattia * p_malattia) + (p_negativo_no_malattia * p_no_malattia)
p_malattia_dato_negativo = (p_negativo_malattia * p_malattia) / p_negativo

print(f"Probabilità a posteriori di avere la malattia dato risultato negativo: {p_malattia_dato_negativo* 100:.2f} %")
```

Notiamo che:
- $\mathbb{P}(A | B) + \mathbb{P}(A |\neg B)$ non deve essere necessariamente $= 1$  
- $\mathbb{P}(A | B) + \mathbb{P}(\neg A | B) = 1$
- $\mathbb{P}(A | \neg B) + \mathbb{P}(\neg A | \neg B) = 1$  

**Corollario**

Sia $\{A_1, ..., A_n\}$ una partizione dell'insieme $\Omega$, $\bigcup\limits_{i=1}^{n} A_{i} = \Omega \land \forall i, j \ | \ i \neq j \rightarrow A_i \cap A_j = \emptyset$

Allora

$$\Large
\mathbb{P}(A_j | B) = \frac{\mathbb{P}(B | A_j) \cdot \mathbb{P}(A_j)}{\sum\limits_{i=1}^{n}\mathbb{P}(B | A_i)\cdot\mathbb{P}(A_i)}
$$


#### Esempio 1
Supponiamo che ci siano tre tipi di giocatori in un torneo di tennis: A, B e C. Il cinquanta per cento dei partecipanti al torneo sono giocatori di tipo A, il 25% sono giocatori di tipo B e il 25% sono giocatori di tipo C. La probabilità di battere i partecipanti dipende dalla classe del giocatore, come segue:

- 0,3 contro un giocatore di tipo A
- 0,4 contro un giocatore di tipo B
- 0,5 contro un giocatore di tipo C

Se giochi una partita in questo torneo, qual è la probabilità di vincere la partita? Supponendo di aver vinto una partita, qual è la probabilità di aver giocato contro un giocatore di tipo A?

<img src="./images/ex1.png">

```python
# Probabilità dei tipi di giocatori
prob_A = 0.50
prob_B = 0.25
prob_C = 0.25

# Probabilità di vincere contro ciascun tipo di giocatore
prob_vincita_A = 0.3
prob_vincita_B = 0.4
prob_vincita_C = 0.5

# Calcolare la probabilità totale di vincere la partita usando il teorema della probabilità totale
prob_vincita_totale = (prob_A * prob_vincita_A) + (prob_B * prob_vincita_B) + (prob_C * prob_vincita_C)

# Probabilità di giocare contro un giocatore di tipo A dato che hai vinto la partita
prob_A_dato_vincita = (prob_A * prob_vincita_A) / prob_vincita_totale

print("La probabilità di vincere la partita è:", round(prob_vincita_totale*100, 2), "%")
print("La probabilità di aver giocato contro un giocatore di tipo A dato che hai vinto è:", round(prob_A_dato_vincita*100, 4), "%")
```
